#### <font color = blue> Here we are removing outliers from 3 Numerical variables like AMT_CREDIT ,AMT_ANNUITY ,AMT_APPLICATION  </font>


```python
Pre_App_DF_AMT=Pre_App_DF.loc[:,['AMT_CREDIT','AMT_APPLICATION']]
```

```python
Pre_App_DF_AMT.isnull().sum()
```

```python
sns.boxplot(data=Pre_App_DF_AMT)
```


#### <font color = blue>  AS you can see from boxplots there are lot of outliers in 3 of Numerical variable we will be remomving them by using The interquartile range (IQR), also called the midspread or middle 50% </font>

```python
Q1 = Pre_App_DF_AMT.quantile(0.5)
Q3 = Pre_App_DF_AMT.quantile(0.95)
IQR = Q3 - Q1
Pre_App_AMT_out = Pre_App_DF_AMT[~((Pre_App_DF_AMT < (Q1 - 1.5 * IQR)) |(Pre_App_DF_AMT > (Q3 + 1.5 * IQR))).any(axis=1)]
Pre_App_AMT_out.shape
```

```python
Pre_App_AMT_out.head()
```

```python
sns.boxplot(data=Pre_App_AMT_out)
```

#### <font color = blue>  AS you can see from boxplots we have removed most of the outliers so we can do statistical analysis </font>

```python
corre=Pre_App_AMT_out.corr(method='pearson')
corre
```

#### <font color = blue>  AS you can see from There is Positive correlation matrix between this two
</font>

###  Bivariate anaylysis with Numerical varible vs Continous Variable
</font>

```python
Pre_App_AMT_out1 = pd.concat([Pre_App_DF.loc[:,'NAME_CONTRACT_STATUS'], Pre_App_AMT_out], axis=1)
```

```python

Pre_App_AMT_out2=Pre_App_AMT_out1.dropna()
```

```python
Pre_App_AMT_out2.index = pd.RangeIndex(len(Pre_App_AMT_out2.index))
Pre_App_AMT_out2.head()
```

```python
Pre_App_AMT_out3=Pre_App_AMT_out2.groupby(['NAME_CONTRACT_STATUS'])
Pre_App_AMT_out3.head()
```

```python
Pre_App_AMT_out4=Pre_App_AMT_out3['AMT_CREDIT','AMT_APPLICATION'].mean()
Pre_App_AMT_out4.head()
```

```python
Pre_App_AMT_out4.plot(kind="bar")

plt.show()
```

#### <font color = red> Refused rate is greater for greater final loan amount but approval rate is moderate for average Loan Amount  </font>
